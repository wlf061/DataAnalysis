# scatter plot 散点图 描述两个变量之间的关系

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
##数据准备
N=1000
x=np.random.randn(N)
y=np.random.randn(N)
#用matplotlib 画散点图
plt.scatter(x,y,marker='x')
plt.show()
#用seaborn 画散点图
df=pd.DataFrame({'x':x,'y':y})
sns.jointplot(x='x',y='y',data=df,kind='scatter'); ##x,y 是data的下标
plt.show()
```

# 折线图 ：表示数据随着时间变化的趋势

```python
### 在 Matplotlib 中 直接使用 plt.plot 函数，需要提前把数据按照x 轴的 大小进行
### 排序 
### 在seaborn 中， 使用 sns.lineplot 函数
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#数据准备
x=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
y=[5,3,6,20,17,16,19,30,32,35]
#使用matplotlib 画折线图
plt.plot(x,y)
plt.show()
#使用seaborn 画折线图
df = pd.DataFrame({'x':x,'y':y})
sns.lineplot(x='x',y='y',data=df)
plt.show()
```

# 直方图 可以看到变量的数值分布

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# 数据准备
a = np.random.randn(100)
s = pd.Series(a) 
# 用 Matplotlib 画直方图
plt.hist(s)
plt.show()
# 用 Seaborn 画直方图
sns.distplot(s, kde=False)
plt.show()
sns.distplot(s, kde=True)
plt.show()
```

# 条形图 可用于查看类别的特征， 在条形图中，长条形的长度表示类别的频数， 宽度表示类别 

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# 数据准备
x = ['Cat1', 'Cat2', 'Cat3', 'Cat4', 'Cat5']
y = [5, 4, 8, 12, 7]
# 用 Matplotlib 画条形图
plt.bar(x, y)
plt.show()
# 用 Seaborn 画条形图
sns.barplot(x, y)
plt.show()
```

# 箱线图 由5个数值点组成：max,min, median, 上下四分位数（Q3,Q1）
### matplotlib 中 使用 plt.boxplot
### seaborn 中 使用 sns.boxplot(x=None, y=None,data=None)函数。 其中data 为DataFrame, x,y 为data 中的变量

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# 数据准备
# 生成 0-1 之间的 10*4 维度数据
data=np.random.normal(size=(10,4)) 
lables = ['A','B','C','D']
# 用 Matplotlib 画箱线图
plt.boxplot(data,labels=lables)
plt.show()
# 用 Seaborn 画箱线图
df = pd.DataFrame(data, columns=lables)
sns.boxplot(data=df)
plt.show()
```

# 饼图 pie

```python
import matplotlib.pyplot as plt
%matplotlib inline
# 数据准备
nums = [25, 37, 33, 37, 6]
labels = ['High-school','Bachelor','Master','Ph.d', 'Others']
# 用 Matplotlib 画饼图
plt.pie(x = nums, labels=labels)
plt.show()
```

# 热力图

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# 数据准备
flights = sns.load_dataset("flights")
data=flights.pivot('year','month','passengers')
# 用 Seaborn 画热力图
sns.heatmap(data)
plt.show()
```
